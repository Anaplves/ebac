# **Exercícios**

## 1\. Pinguins

Neste exercício, vamos utilizar uma base de dados com informações sobre penguins.  
A idéia é preparar a base de dados para prever a espécie do penguin (variável resposta) baseado em suas características físicas e geográficas (variáveis preditivas).

```
#import numpy as np
#import pandas as pd
#import seaborn as sns

from numpy.random import choice as np_random_choice
from pandas import Series as pd_Series
from pandas import concat as pd_concat
from pandas import get_dummies as pd_get_dummies
from seaborn import load_dataset as sns_load_dataset
```

```
# Carregar o dataset "penguins" atribuindo-o para uma variável
ds_penguins = sns_load_dataset('penguins')
```

```
#  Exibir os 5 primeiros registros
ds_penguins.head()
```

### **1.1. Valores nulos**

A base de dados possui valores faltantes, utilize os conceitos da aula para trata-los.

```
# Campos que têm valores nulos:
# - bill_length_mm
# - bill_depth_mm
# - flipper_length_mm
# - body_mass_g
# - sex

# Para os valores numéricos, que são todos os campos previamente citados, exceto
# "sex", seus valores serão preenchidos com sua média.
media_bill_length_mm = ds_penguins['bill_length_mm'].mean()
media_bill_depth_mm = ds_penguins['bill_depth_mm'].mean()
media_flipper_length_mm = ds_penguins['flipper_length_mm'].mean()
media_body_mass_g = ds_penguins['body_mass_g'].mean()

# Preencher os valores nulos com as médias
ds_penguins['bill_length_mm'].fillna(media_bill_length_mm, inplace=True)
ds_penguins['bill_depth_mm'].fillna(media_bill_depth_mm, inplace=True)
ds_penguins['flipper_length_mm'].fillna(media_flipper_length_mm, inplace=True)
ds_penguins['body_mass_g'].fillna(media_body_mass_g, inplace=True)
```

```
# Verificando os 10 primeiros registros do dataset
ds_penguins.head(10)
```

Nota-se que o campo "`sex`" ainda tem valores nulos...  
Para preencher esses espaços será utilizada a proporção dos valores pré existentes.

```
# Cálculo da distribuição de frequência da coluna "sex"
# Aqui tem-se a proporção dos valores
distribuicao_sex = ds_penguins["sex"].value_counts(normalize=True)

# Array de valores da coluna
arr_sex = distribuicao_sex.index

# Tamanho do dataset
ds_len = len(ds_penguins)

# Valores da distribuição dos valores
dist_values = distribuicao_sex.values

# Valores do campo sex
sex_values = np_random_choice(a=arr_sex, size=ds_len, p=dist_values)

# Pandas Series com os valores preenchidos do campo sex
ps = pd_Series(sex_values)

# Preencher os valores nulos da coluna sex com os valores da Pandas Series
ds_penguins['sex'].fillna(ps, inplace=True)

# Verificando os 10 primeiros registros do dataset
ds_penguins.head(10)
```

### **1.2. Variáveis numéricas**

Identifique as variáveis numéricas e crie uma nova coluna **padronizando** seus valores.  
A nova coluna deve ter o mesmo nome da coluna original acrescidade de "*_std*".

> **Nota**: Você não deve tratar a variável resposta.

```
# Tupla com os nomes das colunas a serem padronizadas
c = ('bill_length_mm', 'bill_depth_mm', 'flipper_length_mm', 'body_mass_g')

# Criação de função para automatizar a tarefa
def std_columns(columns: tuple) -> None:
  '''
  Função para padronizar colunas
  '''

  # Loop sobre os valores da tupla
  for i in columns:
    # Nome da coluna padronizada
    std_column = f'{i}_std'
    # Média
    mean_ = ds_penguins[i].mean()
    # Desvio padrão
    std_ = ds_penguins[i].std()

    # Coluna padronizada bill_length_mm_std
    ds_penguins[std_column] = ds_penguins[i].apply(
        lambda x: (x - mean_) / std_
        )

# Executaar a função
std_columns(c)

# Visualizar os 5 primeiros registros
ds_penguins.head()
```

### **1.3. Variáveis categóricas**

Identifique as variáveis categóricas nominais e ordinais.  
Crie uma nova coluna aplicando a técnica correta de conversão a seus valores.  
A nova coluna deve ter o mesmo nome da coluna original acrescidade de "*_nom*" ou "*_ord*".

> **Nota**: Você não deve tratar a variável resposta.

```
# Aplicar one-hot encoding usando get_dummies
cols = ['island', 'sex']

# Um novo dataset somente com as colunas a serem tratadas
ds_penguins_selected = ds_penguins[cols]

ds_penguins_selected = pd_get_dummies(
    ds_penguins_selected,
    columns=cols,
    prefix=cols,
    prefix_sep='_',

    )

# Adicionar o sufixo _nom às colunas
ds_penguins_selected = ds_penguins_selected.add_suffix('_nom')

# Visualizar os 5 primeiros registros
ds_penguins_selected.head()
```

```
# Juntar os dois datasets
ds_penguins = pd_concat([ds_penguins, ds_penguins_selected], axis=1)

# Eliminar o dataset que não é mais necessário
del ds_penguins_selected

# Visualizar os 5 primeiros registros
ds_penguins.head()
```

### **1.4. Limpeza**

Descarte as colunas originais e mantenha apenas a variável resposta e as variáveis preditivas com o sufixo *_std*", *_nom*" e "*_ord*".

```
# Colunas para serem excluídas
cols_drop = [
    'island',
    'bill_length_mm',
    'bill_depth_mm',
    'flipper_length_mm',
    'sex',]

# Excluir as colunas
ds_penguins = ds_penguins.drop(columns=cols_drop)

# Exibir os 5 primeiros registros do dataset
ds_penguins.head()
```
